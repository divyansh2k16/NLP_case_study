### Import necessary modules

### Doing some word and sentence counting

```python
# Count the words
df['tokens'] = df['content'].progress_apply(lambda x: len(nlp(x)))
```

```python
# Count the sentences
df['sentences'] = df['content'].progress_apply(lambda x: len(list(nlp(x).sents)))
```

### Some Basic Sentiment Analysis

```python
# using textblob to do the sentiment analysis - because it's faster than spacy
df['sentiment'] = df['content'].progress_apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
# Save as CSV
df.to_csv('outputs/preprocessed/brewery_reviews.csv')
```

```python
df_reviews = pd.read_csv('outputs/preprocessed/brewery_reviews.csv', parse_dates=['date'], index_col=0)
df_reviews[[col for col in df_reviews.columns.unique() if 'id' not in col]].describe()
```

```python
# What is the percent of reviews that are wordless
len(df_reviews[df_reviews['content'].isnull()]) / len(df_reviews)
```

```python
# checking the number of reviews without any tokens and dropping them in case there are any
if df_reviews.tokens.isnull().sum() > 0:
    df_reviews.dropna(inplace=True)
```

```python
sns.set_style("darkgrid", {"axes.facecolor": ".925"})
sns.displot(df_reviews, x='tokens', color="tab:green", aspect=2.5)
plt.xlabel('Number of Tokens per Review')
plt.ylabel('Count')
plt.title('Distribution of Tokens per Review')
plt.savefig('outputs/figures/distribution-tokens-per-review.png', bbox_inches='tight')
plt.show()
```

```python
sns.displot(data=df_reviews, x='sentences', aspect=2.5, color="tab:green", kind="hist", discrete=True)
plt.xlabel('Number of Sentences per Review')
plt.ylabel('Count')
plt.title('Distribution of Sentences per Review')
plt.savefig('outputs/figures/distribution-sentences-per-review.png', bbox_inches='tight')
plt.show()
```

```python
sns.displot(data=df_reviews, x='sentiment', aspect=2.5, color="tab:green")
plt.xlabel('Sentiment per Review')
plt.ylabel('Reviews')
sns.despine()
plt.show()
```

> Wow! The sentiments are almost perfectly normally distributed!

```python
# Check the number of tokens
num_tokens = df_reviews['tokens'].sum()
print (f"There are {num_tokens} tokens.")

# Check the number of raw "words"
text = " ".join(review for review in df_reviews.content)
print (f"There are {len(text)} words.")
```

```python
# Generate and save a word cloud image
wordcloud = WordCloud(background_color="white").generate(text)
wordcloud.to_file('outputs/figures/word-cloud-brewery-reviews.png')

# Show image
plt.figure(figsize=(15,20))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
